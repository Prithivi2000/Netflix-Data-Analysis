# Netflix Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

# Importing Dataset and removing unwanted columns

```python
dataset = pd.read_csv('netflix_titles.csv')
```

```python
dataset = dataset.drop('cast',axis=1)
# dataset = dataset.drop('director',axis = 1)
dataset = dataset.drop('show_id',axis=1)
dataset = dataset.drop('description',axis=1)
dataset
```

# Top 10 Countries whose name is repeated  

```python
data_about_country = pd.DataFrame(dataset['country'].value_counts())
data_about_country = data_about_country.head(10)
data_about_country.rename(columns={'country':'Appeared times'}, inplace = True)
data_about_country
```

## Graph Displaying the frequency of the top 10 countries

```python
x=[]
y=[]
for i in data_about_country.index:
    x.append(i)
    y.append(data_about_country['Appeared times'][i])
plt.figure(figsize=(13,10))
plt.bar(x,y,color='rgb')
plt.xlabel('Country')
plt.ylabel('Appeared in Dataset')
plt.title('Countries Having high frequency in dataset',fontsize= 40)
plt.show()
```

# Visualising the Category(Movies or Tv Shows) which is available more

```python
dataset['type'].value_counts()
x_type= ['Movies', 'TV Show']
y_type= [4265, 1969]
plt.figure(figsize=(13,10))
plt.bar(x_type,y_type,color='rgb')
plt.xlabel('Avalability', fontsize=18)
plt.ylabel('Genre',fontsize=18)
plt.title('Category Vs Availability',fontsize= 40)
plt.show()
```

# Visualising the top 5 directors in Netflix

```python
data_about_director = pd.DataFrame(dataset['director'].value_counts())
data_about_director = data_about_director.head(5)
x_director =[]
y_director =[]
for i in data_about_director.index:
    x_director.append(i)
    y_director.append(data_about_director['director'][i])

plt.figure(figsize=(13,10))
plt.bar(x_director,y_director,color='rgb')
plt.xlabel('Director', fontsize=18)
plt.ylabel('No. of Films Added', fontsize=18)
plt.title('Famous Directors in Netflix',fontsize= 40)
plt.show()
```

# Visualising which duration movie or tv show is available by how much quantity

```python
data_duration = pd.DataFrame(dataset.duration.value_counts())
data_duration = data_duration.head(20)
x_duration=[]
y_duration=[]
for i in data_duration.index:
    x_duration.append(i)
    y_duration.append(data_duration['duration'][i])
plt.figure(figsize=(25,10))
plt.plot(x_duration,y_duration,color='r')
plt.xlabel('Type', fontsize=15)
plt.ylabel('Duration', fontsize=15)
plt.title('Duaration of Shows',fontsize= 40)
plt.show()
```
